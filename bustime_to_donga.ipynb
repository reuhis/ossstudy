```python
# prompt: /content/bus_information/주요노선_평균_이용인원_및_통행량_20250320000959.csv
# 해당 자료의 열 중 부산에 해당하는 값만 기억하고 나머지는 무시해줘
# 그리고 자료의 행에 해당하는 것 중 평균 이용인원은 하루 일에 이용한 사람들의 이용인원을 나타낸 것이야 즉, 인/일 이야
# 행에 해당하는 것 중 평균 통행량은 하루 일에 이용한 통행량을 나타낸 것이야 즉, 통행량/일 이야
# 행 부산에서 구분된 일주일, 주중 같은 값은 해당하는 날의 특징을 구분해서 각 날마다 해당되는 평균 통행량과 평균 이용인원을 나타낸 것이야
# 내가 말한 것을 나중에 쓰기 위해서 기억해줘

import pandas as pd

# 파일 경로 설정
file_path = '/content/bus_information/주요노선_평균_이용인원_및_통행량_20250320000959.csv'

# CSV 파일 읽어오기 (cp949 인코딩으로 시도)
try:
  df = pd.read_csv(file_path, encoding='cp949')
except UnicodeDecodeError:
  # cp949 인코딩이 실패하면 utf-8 인코딩으로 시도
  df = pd.read_csv(file_path, encoding='utf-8')

# '부산' 열 추출
busan_data = df[df['구분'] == '부산']

# '일주일', '주중' 등 날짜/요일 특징과 '평균 이용인원', '평균 통행량' 정보를 저장
busan_daily_info = {}
for index, row in busan_data.iterrows():
  day_feature = row['일자/요일']
  avg_passengers = row['평균 이용인원']
  avg_traffic = row['평균 통행량']

  if day_feature not in busan_daily_info:
    busan_daily_info[day_feature] = {'평균 이용인원': [], '평균 통행량': []}
  busan_daily_info[day_feature]['평균 이용인원'].append(avg_passengers)
  busan_daily_info[day_feature]['평균 통행량'].append(avg_traffic)

# 저장된 정보 출력 (선택 사항)
for day_feature, info in busan_daily_info.items():
  print(f"{day_feature}:")
  print(f"  평균 이용인원: {info['평균 이용인원']}")
  print(f"  평균 통행량: {info['평균 통행량']}")
  print()

# 나중에 사용하기 위해 정보 저장
busan_average_passenger_and_traffic = busan_daily_info
```

```python
# prompt: 시도별_대중교통_접근시간_20250319142710.csv
# 시도별_환승소요시간_20250319142941.csv
# 한국교통안전공단_전국 대중교통 버스 평균 운행속도_20231231.csv
# 한국교통안전공단_전국 지역별 대중교통 이용 및 환승실태_20231231.csv
# 위의 자료들의 접근시간, 환승시간, 버스운행시간 등을 분석하고
# 사용자로부터 자신의 위치에서 동아대까지의 대략적인 거리를 입력받아서 주어진 거리에 대해 얼마나 시간이 걸릴지 상세하게 분석해줘

import pandas as pd

# 사용자로부터 거리를 입력받습니다.
distance_to_dong_a = float(input("동아대학교까지의 거리를 입력해주세요 (km): "))

# CSV 파일 경로 (예시)
file_path_access_time = '/content/bus_information/시도별_대중교통_접근시간_20250319142710.csv'
file_path_transfer_time = '/content/bus_information/시도별_환승소요시간_20250319142941.csv'
file_path_bus_speed = '/content/bus_information/한국교통안전공단_전국 대중교통 버스 평균 운행속도_20231231.csv'

# 접근 시간 데이터 불러오기
try:
  df_access_time = pd.read_csv(file_path_access_time, encoding='cp949')
except UnicodeDecodeError:
  df_access_time = pd.read_csv(file_path_access_time, encoding='utf-8')

# 환승 시간 데이터 불러오기
try:
  df_transfer_time = pd.read_csv(file_path_transfer_time, encoding='cp949')
except UnicodeDecodeError:
  df_transfer_time = pd.read_csv(file_path_transfer_time, encoding='utf-8')

# 버스 평균 운행 속도 데이터 불러오기
try:
  df_bus_speed = pd.read_csv(file_path_bus_speed, encoding='cp949')
except UnicodeDecodeError:
  df_bus_speed = pd.read_csv(file_path_bus_speed, encoding='utf-8')


# 부산의 평균 접근 시간 추출 (예시)
average_access_time_busan = df_access_time['부산'].mean()

# 부산의 평균 환승 시간 추출 (예시)
average_transfer_time_busan = df_transfer_time['부산'].mean()

# 부산의 평균 버스 운행 속도 추출 (예시)
average_bus_speed_busan = df_bus_speed[df_bus_speed['시도'] == '부산']['평균 운행속도'].mean()

# 대략적인 이동 시간 계산 (단순화된 계산)
estimated_travel_time = (distance_to_dong_a / average_bus_speed_busan) * 60 + average_access_time_busan + average_transfer_time_busan

print(f"\n동아대학교까지의 대략적인 이동 시간은 {estimated_travel_time:.2f}분 정도 소요될 것으로 예상됩니다.")

# 추가 분석 (시간대별/요일별 혼잡도 고려)
print("\n추가 분석:")
print("- 출퇴근 시간대(오전 7시~9시, 오후 5시~7시)에는 버스 이용객이 많아 이동 시간이 더 소요될 수 있습니다.")
print("- 평일보다 주말에는 이용객이 적어 이동 시간이 다소 단축될 수 있습니다.")
print("- 특정 노선이나 버스 유형에 따라 이동 시간이 달라질 수 있습니다.")
```

```python
# prompt: 이제 주석 '자료를 분석한 결과' 밑에 있는 코드를 이용해서 사용자로부터 자신의 위치에서 동아대 까지의 대략적인 거리를 입력받아 해당 거리를 버스를 이용해서 얼마나 시간이 걸릴지 계산하는 코드를 작성해줘

# 사용자로부터 거리를 입력받습니다.
distance_to_dong_a = float(input("동아대학교까지의 거리를 입력해주세요 (km): "))

# 버스 평균 속도 (km/h) - 실제 데이터를 사용하는 것이 좋습니다.
average_bus_speed = 20  # 예시 값, 실제 데이터로 변경해야 합니다.

# 대략적인 이동 시간 계산 (분)
estimated_travel_time = (distance_to_dong_a / average_bus_speed) * 60

print(f"\n동아대학교까지 버스를 이용해서 이동하는 데 대략 {estimated_travel_time:.2f}분 정도 소요될 것으로 예상됩니다.")

# 추가 분석 (시간대별/요일별 혼잡도 고려)
print("\n추가 분석:")
print("- 출퇴근 시간대(오전 7시~9시, 오후 5시~7시)에는 버스 이용객이 많아 이동 시간이 더 소요될 수 있습니다.")
print("- 평일보다 주말에는 이용객이 적어 이동 시간이 다소 단축될 수 있습니다.")
print("- 특정 노선이나 버스 유형에 따라 이동 시간이 달라질 수 있습니다.")
```

```python
# prompt: folium을 사용해서 위도35.116657, 경도128.968494 에서 주석 '최종결과'에서 입력받은 값(km)만큼 떨어진 랜덤한 곳으로부터 해당 위치까지의 오는길을 표시해줘

!pip install folium
import folium
import random

# 최종결과에서 입력받은 값(km)을 가져옵니다.
distance_km = float(input("최종결과에서 입력받은 값(km)을 입력하세요: "))

# 동아대학교 좌표 (위도, 경도)
dong_a_lat = 35.116657
dong_a_lon = 128.968494

# 랜덤한 지점 생성
def generate_random_point(lat, lon, distance_km):
  """주어진 좌표에서 주어진 거리(km)만큼 떨어진 랜덤한 지점을 생성합니다."""
  radius_degree = distance_km / 111.32  # 대략적인 1도의 거리 (km)
  random_bearing = random.uniform(0, 360)  # 0~360도 사이의 임의의 방위각

  # 랜덤한 지점의 위도, 경도 계산
  random_lat = lat + radius_degree * random.uniform(-1, 1)
  random_lon = lon + radius_degree * random.uniform(-1, 1)

  return random_lat, random_lon

# 랜덤한 지점 생성
random_lat, random_lon = generate_random_point(dong_a_lat, dong_a_lon, distance_km)

# 지도 생성
m = folium.Map(location=[dong_a_lat, dong_a_lon], zoom_start=13)

# 출발점과 도착점 마커 추가
folium.Marker(location=[random_lat, random_lon], popup='출발점', icon=folium.Icon(color='blue')).add_to(m)
folium.Marker(location=[dong_a_lat, dong_a_lon], popup='동아대학교', icon=folium.Icon(color='red')).add_to(m)

# 출발점에서 도착점까지의 경로 표시 (polyline)
folium.PolyLine(locations=[[random_lat, random_lon], [dong_a_lat, dong_a_lon]], color='blue').add_to(m)


# 지도 출력
m
```
